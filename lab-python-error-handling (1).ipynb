````python
#Step 1: Initialize Inventory with Error Handling


def initialize_inventory(products):
    inventory = {} 
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
                inventory[product] = quantity  # Store valid quantity
            except ValueError as error:
                print(f"Error: {error}")
    return inventory

````

````python
#Step 2: Calculate Total Price with Error Handling

def calculate_total_price(products):
    prices = {} 
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative. Please enter a valid price.")
                valid_price = True
                prices[product] = price  # Store valid price
            except ValueError as error:
                print(f"Error: {error}")
    return prices
   
````

````python
#Step 3: Get Customer Orders with Error Handling


def get_customer_orders(inventory):
    orders = {}  # Dictionary to store customer orders
    while True:
        product = input("Enter the product name (or 'exit' to finish): ").lower()
        if product == "exit":
            break
        if product not in inventory:
            print("Error: Product not found in inventory. Please enter a valid product.")
            continue
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity for {product}: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")
                if quantity > inventory[product]:
                    print(f"Error: Not enough stock available. Only {inventory[product]} left.")
                else:
                    valid_quantity = True
                    orders[product] = quantity  # Store valid order
            except ValueError as error:
                print(f"Error: {error}. Please enter a valid quantity.")
    return orders
````

````python
#Step 4: Run the Full Program

def main():
    products = ["cow", "chicken", "sheep", "goat"]  
    
    print("\nInitializing Inventory:")
    inventory = initialize_inventory(products) 
    
    print("\nSetting Product Prices:")
    prices = calculate_total_price(products)  

    print("\nCustomer Orders:")
    customer_orders = get_customer_orders(inventory)  

    print("\nFinal Data:")
    print("Inventory:", inventory)
    print("Prices:", prices)
    print("Customer Orders:", customer_orders)

# Run the program
main()

````
